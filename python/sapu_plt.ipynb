```python
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
plt.rcParams["font.size"] =14
fig,ax = plt.subplots(facecolor="white")
ax.plot(["月","火","水","木","金","土","日"],[19,22,24,21,20,22,20],label="最高気温",marker="o")
ax.bar(["月","火","水","木","金","土","日"],[19,22,24,21,20,22,20],label="最高気温")
ax.plot(["月","火","水","木","金","土","日"],[10,13,12,12,10,11,12],label="最低気温",marker="^",color="m")
ax.set_xlabel("曜日")
ax.set_ylabel("気温",rotation="horizontal")
ax.set_yticks([0,5,10,15,20,25])
ax.set_title("東京　10月1周目")
ax.grid()
ax.legend()
```
